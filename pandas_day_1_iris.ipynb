##### 1. Write a Pandas program to get the data types of the given csv data (iris.csv ) fields

```python
import pandas as pd

# Load the iris dataset from CSV file
file_path = 'iris.csv'
df = pd.read_csv(file_path)

# Display the data types of each column
data_types = df.dtypes
print(data_types)
```

##### 2. Write a Pandas program to find the sum, mean, max, min value of 'SepalLengthCm' column of dataframe

```python
# Extract 'SepalLengthCm' column
sepal_length_column = df['SepalLengthCm']

# Calculate and print the sum, mean, max, and min values
sum_value = sepal_length_column.sum()
mean_value = sepal_length_column.mean()
max_value = sepal_length_column.max()
min_value = sepal_length_column.min()

print(f"Sum: {sum_value}")
print(f"Mean: {mean_value}")
print(f"Max: {max_value}")
print(f"Min: {min_value}")
```

##### 3. Write a Pandas program to import iris data and skipping first twenty rows into a Pandas dataframe.

```python
import pandas as pd

# Load the iris dataset from CSV file, skipping the first twenty rows
file_path = 'iris.csv'
df = pd.read_csv(file_path, skiprows=range(1, 21))

# Display the resulting DataFrame
df
```

##### 4. Write a Pandas program to display the last ten rows.

```python
import pandas as pd

# Load the iris dataset from CSV file
file_path = 'iris.csv'
df = pd.read_csv(file_path)

# Display the last ten rows of the DataFrame
last_ten_rows = df.tail(10)
last_ten_rows
```

##### 5. Write a Pandas program to create a subtotal of "PetalLengthCm" against Species from the dataframe.

```python
# Create a subtotal of "PetalLengthCm" against "Species"
subtotal = df.groupby('Species')['PetalLengthCm'].sum()

# Display the subtotal
print(subtotal)
```

###### 6. Write a Pandas program to find details where "PetalLengthCm" > 2.

```python
# Find details where "PetalLengthCm" > 2
filtered_data = df[df['PetalLengthCm'] > 2]

# Display the filtered DataFrame
filtered_data
```

###### 7. Write a Pandas program find all records that include two specific Species.

```python
# Define the two specific species
target_species = ['setosa', 'versicolor']

# Find all records that include the two specific species
filtered_data = df[df['Species'].isin(target_species)]

# Display the filtered DataFrame
filtered_data
```

###### 8.  Write a Pandas program to sort the records by the SepalLengthCm column.

```python
# Sort the records by the "SepalLengthCm" column
sorted_data = df.sort_values(by='SepalLengthCm')

# Display the sorted DataFrame
sorted_data
```

##### 9. Write a Pandas program to insert a column in the sixth position of dataframe and fill it with NaN values.

```python
'''
It is impossible to insert a column in the sixth position of dataframe and fill it with NaN values,
because there is a column named Species already present at sixth position, 
we can insert new column filled with NaN values in the fifth position
'''

# Insert a new column filled with NaN values in the fifth position
new_column_name = 'NewColumn'
df.insert(5, new_column_name, value=pd.NA)

# Display the DataFrame with the new column
df
```

##### 10. Write a Pandas program to add summation to a row of the dataframe.

```python
# Calculate the sum of each column
sum_row = df.sum(numeric_only=True)

# Add the sum row to the DataFrame
df.loc['Sum'] = sum_row

# Display the DataFrame with the sum row
df
```
