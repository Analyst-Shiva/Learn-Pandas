```python
import pandas as pd
import numpy as np
data  = {'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],
        'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],
        'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
        'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']}
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']
```

##### Write a Pandas program to create a dataframe by given dictionary and index as lables

```python
# Create a DataFrame using the provided data and labels
df = pd.DataFrame(data, index=labels)

# Display the DataFrame
df
```

###### import the titanic dataset

```python
# Load the Titanic dataset from CSV file
file_path = 'titanic.csv'
df = pd.read_csv(file_path)

# Display the DataFrame
df
```

##### 1. Write a Pandas program to create a Pivot table and find the total fare amount class wise, gender wise.

```python
# Create a Pivot table to find the total fare amount class-wise and gender-wise
pivot_table_result = df.pivot_table(values='fare', index=['pclass', 'sex'], aggfunc='sum')

# Display the Pivot table
pivot_table_result
```

##### 2. Write a Pandas program to create a Pivot table and find survival rate by gender.

```python
# Create a Pivot table to find the survival rate by gender
pivot_table_result = df.pivot_table(values='survived', index='sex', aggfunc='mean')

# Display the Pivot table
pivot_table_result
```

###### 3. Write a Pandas to create to find the total count no of people survived pclass wise, class wise and gender wise.

```python
# Create a Pivot table to find the total count of people survived by Pclass, class, and gender
pivot_table_result = df.pivot_table(values='survived', index=['pclass', 'class', 'sex'], aggfunc='count')

# Display the Pivot table
pivot_table_result
```

###### 4. Write a Pandas program to create a Pivot table and calculate how many women and men were in a particular cabin class.

```python
# Create a Pivot table to calculate the count of women and men in each cabin class
pivot_table_result = df.pivot_table(values='survived', index='pclass', columns='sex', aggfunc='count')

# Display the Pivot table
pivot_table_result
```

##### 5. Write a Pandas program to count the number of missing values in each column.

```python
# Count the number of missing values in each column
missing_values_count = df.isnull().sum()

# Display the count of missing values
missing_values_count
```

##### 6. Write a Pandas program to replace null values with the value from the previous row or the next row in a given DataFrame.

```python
# Replace null values with the value from the previous row (forward fill)
df_filled_forward = df.fillna(method='ffill')

# Replace null values with the value from the next row (backward fill)
df_filled_backward = df.fillna(method='bfill')

# Display the original DataFrame
print("Original DataFrame:")
print(df)

# Display the DataFrame after forward fill
print("\nDataFrame after forward fill:")
print(df_filled_forward)

# Display the DataFrame after backward fill
print("\nDataFrame after backward fill:")
print(df_filled_backward)
```

###### 7. Write a Pandas program to replace nulls, with median or mean of the specified columns in a given DataFrame.

```python
# Specify the columns for which you want to replace null values
columns_to_fill_mean = ['age', 'fare']  # Replace with your actual column names
columns_to_fill_median = ['age', 'fare']  # Replace with your actual column names

# Replace null values with the mean of specified columns
df[columns_to_fill_mean] = df[columns_to_fill_mean].fillna(df[columns_to_fill_mean].mean())

# Replace null values with the median of specified columns
df[columns_to_fill_median] = df[columns_to_fill_median].fillna(df[columns_to_fill_median].median())

# Display the updated DataFrame
df
```

###### 9. Write a Pandas program to remove the duplicates from 'embark_town' column of titanic dataset.

```python
# Remove duplicates from 'embark_town' column
df_unique_embark_town = df.drop_duplicates(subset='embark_town')

# Display the DataFrame after removing duplicates
print("DataFrame after removing duplicates from 'embark_town' column:")
print(df_unique_embark_town[['embark_town']])
```

###### 10. Write a Pandas program to find out the survived is 1 details by the 'Southampton' in who is the 'child' from dataset.

```python
# Filter rows where 'survived' is 1, 'embark_town' is 'Southampton', and 'who' is 'child'
survived_southampton_child = df[(df['survived'] == 1) & (df['embark_town'] == 'Southampton') & (df['who'] == 'child')]

# Display the filtered DataFrame
print("Details of passengers who survived (survived=1) in 'Southampton' and are 'child':")
survived_southampton_child
```

###### 12. Write a Pandas program to filter those records where embarked matches with multiple values (C, Q) from the dataset.

```python
# Filter records where 'embarked' matches with multiple values ('C' or 'Q')
filtered_records = df[df['embarked'].isin(['C', 'Q'])]

# Display the filtered DataFrame
print("Records where 'embarked' matches with 'C' or 'Q':")
filtered_records
```
