```python
import pandas as pd
import numpy as np
data  = {'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],
        'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],
        'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
        'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']}
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']
```

##### Write a Pandas program to create a dataframe by given dictionary and index as lables

```python
# Create a Pandas DataFrame
df = pd.DataFrame(data, index=labels)

# Display the DataFrame
df
```

##### 1. Write a Pandas program to sort the DataFrame first by 'name' in descending order, then by 'score' in ascending order

```python
# Sort the DataFrame first by 'name' in descending order, then by 'score' in ascending order
sorted_df = df.sort_values(by=['name', 'score'], ascending=[False, True])

# Display the sorted DataFrame
sorted_df
```

###### 2. Write a Pandas program to replace the 'qualify' column contains the values 'yes' and 'no' with True and False

```python
# Replace values in the 'qualify' column with True and False
df['qualify'] = df['qualify'].replace({'yes': True, 'no': False})

# Display the DataFrame after replacement
df
```

#####3. Write a Pandas program to delete the 'attempts' column from the DataFrame

```python
# Delete the 'attempts' column from the DataFrame
df = df.drop(columns=['attempts'])

# Display the DataFrame after deletion
df
```

##### 4. Write a Pandas program to insert a new column in existing DataFrame.

```python
# Insert a new column 'result' with default values
df['result'] = 'Pending'

# Display the DataFrame with the new column
df
```

##### 5. Write a Pandas program to change the name 'James' to 'Maxwell' in name column of the DataFrame.

```python
# Change the name 'James' to 'Maxwell' in the 'name' column
df['name'] = df['name'].replace('James', 'Maxwell')

# Display the DataFrame after the change
df
```

##### 6. Write a Pandas program to rename columns of a given DataFrame

```python
# Rename columns
df.rename(columns={'name': 'full_name', 'score': 'marks', 'attempts': 'tries', 'qualify': 'eligible'}, inplace=True)

# Display the DataFrame after renaming columns
df
```

##### 7. Write a Pandas program to convert the datatype of a given column (floats to ints).

```python
# Convert the datatype of the 'marks' column from float to int
df['marks'] = df['marks'].astype(int, errors='ignore')

# Display the DataFrame after the conversion
df
```

###### 8.  Write a Pandas program to convert index in a column of the given dataframe. 

```python
# Move the index into a column
df.reset_index(inplace=True)

# Display the DataFrame after moving the index into a column
df
```

##### 9. Write a Pandas program to shuffle a DataFrame rows.

```python
# Shuffle the rows of the DataFrame
shuffled_df = df.sample(frac=1, random_state=42)

# Display the shuffled DataFrame
shuffled_df
```

```python
student_data1 = pd.DataFrame({
        'student_id': ['S1', 'S2', 'S3', 'S4', 'S5'],
         'name': ['Danniella Fenton', 'Ryder Storey', 'Bryce Jensen', 'Ed Bernal', 'Kwame Morin'], 
        'marks': [200, 210, 190, 222, 199]})

student_data2 = pd.DataFrame({
        'student_id': ['S4', 'S5', 'S6', 'S7', 'S8'],
        'name': ['Scarlette Fisher', 'Carla Williamson', 'Dante Morse', 'Kaiser William', 'Madeeha Preston'], 
        'marks': [201, 200, 198, 219, 201]})
```

#####11. Write a Pandas program to join the two given dataframes along rows and assign all data.

```python
# Join the two DataFrames along rows and assign all data
combined_data = pd.concat([student_data1, student_data2], ignore_index=True)

# Display the combined DataFrame
combined_data
```

##### 12. Write a Pandas program to join the two given dataframes along columns and assign all data.

```python
# Join the two DataFrames along columns and assign all data
combined_data = pd.merge(student_data1, student_data2, on='student_id', how='outer')

# Display the combined DataFrame
combined_data
```

##### 13. Write a Pandas program to join the two dataframes using the common column of both dataframes.

```python
# Join the two DataFrames using the common column 'student_id'
combined_data = pd.merge(student_data1, student_data2, on='student_id')

# Display the combined DataFrame
combined_data
```

###### 14. Write a Pandas program to join (left join) the two dataframes using keys from left dataframe only.

```python
# Left join the two DataFrames using the keys from the left DataFrame only
combined_data = pd.merge(student_data1, student_data2, on='student_id', how='left', suffixes=('_left', '_right'))

# Display the combined DataFrame
combined_data
```

###### 15. Write a Pandas program to join two dataframes using keys from right dataframe only.

```python
# Right join the two DataFrames using the keys from the right DataFrame only
combined_data = pd.merge(student_data1, student_data2, on='student_id', how='right', suffixes=('_left', '_right'))

# Display the combined DataFrame
combined_data
```

######16. Write a Pandas program to create the todays date.

###### 17. Write a Pandas program to get the current date, oldest date and number of days between Current date and oldest date of Ufo dataset.

```python
# Convert 'Date_time' column to datetime format
df['Date_time'] = pd.to_datetime(df['Date_time'], errors='coerce')

# Filter out rows with NaT (Not a Time) values
df = df.dropna(subset=['Date_time'])

# Get the current date
current_date = pd.to_datetime('today').date()

# Get the oldest date from the dataset
oldest_date = df['Date_time'].min().date()

# Calculate the number of days between current date and oldest date
days_difference = (current_date - oldest_date).days

# Display the results
print("Current date:", current_date)
print("Oldest date in the dataset:", oldest_date)
print("Number of days between current date and oldest date:", days_difference)
```

###### 18. Write a Pandas program to get all the info of the dataframe between 1950-10-10 and 1960-10-10.

```python
# Filter the DataFrame between '1950-10-10' and '1960-10-10'
filtered_df = df[(df['Date_time'] >= '1950-10-10') & (df['Date_time'] <= '1960-10-10')]

# Display the information of the filtered DataFrame
filtered_df.info()
```
